```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## create synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples= 1000, n_features=2, n_redundant=0,  n_classes=2, n_clusters_per_class=1)
```

```python
pd.DataFrame(X)[0]
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y, palette='Set1')
```

```python
##trains test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
## SVC classifier
from sklearn.svm import SVC
svc = SVC(kernel='linear', C=1.0)
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
## use rbf kernel
svc_rbf = SVC(kernel='rbf', C=1.0)
svc_rbf.fit(X_train, y_train)
y_pred_rbf = svc_rbf.predict(X_test)
print(classification_report(y_test, y_pred_rbf))
print(confusion_matrix(y_test, y_pred_rbf))
```

```python
## scatter plot for y_test and y_pred
plt.figure(figsize=(10, 6))
# Plot true labels
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_test, palette='Set1', alpha=0.5, marker='o', legend='brief')
# Plot predicted labels with different marker and no legend
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_pred, palette='Set2', alpha=0.5, marker='X', legend=False)

plt.title('True vs Predicted Labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
## use polinomial kernel
svc_poly = SVC(kernel='poly', degree=3, C=1.0)
svc_poly.fit(X_train, y_train)
y_pred_poly = svc_poly.predict(X_test)
print(classification_report(y_test, y_pred_poly))
print(confusion_matrix(y_test, y_pred_poly))
```

```python
## scatter plot for y_test and y_pred
plt.figure(figsize=(10, 6))
# Plot true labels
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_test, palette='Set1', alpha=0.5, marker='o', legend='brief')
# Plot predicted labels with different marker and no legend
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_pred, palette='Set2', alpha=0.5, marker='X', legend=False)

plt.title('True vs Predicted Labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
## use sigmoid kernel
svc_sigmoid = SVC(kernel='sigmoid', C=1.0)
svc_sigmoid.fit(X_train, y_train)
y_pred_sigmoid = svc_sigmoid.predict(X_test)
print(classification_report(y_test, y_pred_sigmoid))
print(confusion_matrix(y_test, y_pred_sigmoid))
```

```python
## scatter plot for y_test and y_pred
plt.figure(figsize=(10, 6))
# Plot true labels
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_test, palette='Set1', alpha=0.5, marker='o', legend='brief')
# Plot predicted labels with different marker and no legend
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_pred, palette='Set2', alpha=0.5, marker='X', legend=False)

plt.title('True vs Predicted Labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
## hyperparameter tuning grid search
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2, n_jobs=-1)
grid.fit(X_train, y_train)
print("Best parameters found: ", grid.best_params_)
print("Best score: ", grid.best_score_)

```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
